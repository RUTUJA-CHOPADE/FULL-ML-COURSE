```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)
```

```python
# Create dummy dataset of 100 students
n_students = 100

# Generate student data
student_data = {
    'Student_ID': np.arange(1001, 1001 + n_students),
    'Name': ['Student_' + str(i) for i in range(1, n_students + 1)],
    'Age': np.random.randint(18, 25, n_students),
    'Gender': np.random.choice(['Male', 'Female', 'Other'], n_students),
    'GPA': np.round(np.random.uniform(1.5, 4.0, n_students), 2),
    'Attendance_Percentage': np.random.randint(60, 100, n_students),
    'Math_Score': np.random.randint(40, 100, n_students),
    'Science_Score': np.random.randint(40, 100, n_students),
    'English_Score': np.random.randint(40, 100, n_students),
    'Study_Hours_Per_Week': np.random.uniform(5, 40, n_students).round(1)
}

df = pd.DataFrame(student_data)
print("Dataset created successfully!")
print(f"Shape: {df.shape}")
print("\nFirst 5 rows:")
print(df.head())
```

```python
# Basic Dataset Information
print("=" * 60)
print("DATASET INFORMATION")
print("=" * 60)
print(df.info())
print("\n" + "=" * 60)
print("MISSING VALUES")
print("=" * 60)
print(df.isnull().sum())
print("\n" + "=" * 60)
print("DATA TYPES")
print("=" * 60)
print(df.dtypes)
```

```python
# Descriptive Statistics
print("=" * 60)
print("DESCRIPTIVE STATISTICS")
print("=" * 60)
print(df.describe().round(2))
print("\n" + "=" * 60)
print("CATEGORICAL DATA SUMMARY")
print("=" * 60)
print("\nGender Distribution:")
print(df['Gender'].value_counts())
print("\nAge Distribution:")
print(df['Age'].value_counts().sort_index())
```

```python
# Visualize Numerical Distributions
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Distribution of Numerical Features', fontsize=16, fontweight='bold')

# GPA Distribution
axes[0, 0].hist(df['GPA'], bins=20, color='skyblue', edgecolor='black')
axes[0, 0].set_title('GPA Distribution')
axes[0, 0].set_xlabel('GPA')
axes[0, 0].set_ylabel('Frequency')

# Age Distribution
axes[0, 1].hist(df['Age'], bins=10, color='lightgreen', edgecolor='black')
axes[0, 1].set_title('Age Distribution')
axes[0, 1].set_xlabel('Age')
axes[0, 1].set_ylabel('Frequency')

# Attendance Distribution
axes[0, 2].hist(df['Attendance_Percentage'], bins=15, color='salmon', edgecolor='black')
axes[0, 2].set_title('Attendance Distribution')
axes[0, 2].set_xlabel('Attendance %')
axes[0, 2].set_ylabel('Frequency')

# Math Score Distribution
axes[1, 0].hist(df['Math_Score'], bins=15, color='gold', edgecolor='black')
axes[1, 0].set_title('Math Score Distribution')
axes[1, 0].set_xlabel('Score')
axes[1, 0].set_ylabel('Frequency')

# Science Score Distribution
axes[1, 1].hist(df['Science_Score'], bins=15, color='plum', edgecolor='black')
axes[1, 1].set_title('Science Score Distribution')
axes[1, 1].set_xlabel('Score')
axes[1, 1].set_ylabel('Frequency')

# English Score Distribution
axes[1, 2].hist(df['English_Score'], bins=15, color='lightcoral', edgecolor='black')
axes[1, 2].set_title('English Score Distribution')
axes[1, 2].set_xlabel('Score')
axes[1, 2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Visualize Categorical Data
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Categorical Variables Analysis', fontsize=16, fontweight='bold')

# Gender Distribution
gender_counts = df['Gender'].value_counts()
axes[0].bar(gender_counts.index, gender_counts.values, color=['#FF6B6B', '#4ECDC4', '#95E1D3'], edgecolor='black')
axes[0].set_title('Gender Distribution')
axes[0].set_ylabel('Count')
axes[0].set_xlabel('Gender')

# Age Distribution by Gender
age_gender = df.groupby(['Age', 'Gender']).size().unstack(fill_value=0)
age_gender.plot(kind='bar', ax=axes[1], color=['#FF6B6B', '#4ECDC4', '#95E1D3'], edgecolor='black')
axes[1].set_title('Age Distribution by Gender')
axes[1].set_ylabel('Count')
axes[1].set_xlabel('Age')
axes[1].legend(title='Gender')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Correlation Analysis
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Numerical Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nTop Correlations with GPA:")
gpa_corr = correlation_matrix['GPA'].sort_values(ascending=False)
print(gpa_corr[gpa_corr.index != 'GPA'])
```

```python
# Box Plot Analysis
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Box Plot Analysis by Gender', fontsize=16, fontweight='bold')

# GPA by Gender
sns.boxplot(data=df, x='Gender', y='GPA', ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('GPA by Gender')

# Attendance by Gender
sns.boxplot(data=df, x='Gender', y='Attendance_Percentage', ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('Attendance % by Gender')

# Math Score by Gender
sns.boxplot(data=df, x='Gender', y='Math_Score', ax=axes[1, 0], palette='Set2')
axes[1, 0].set_title('Math Score by Gender')

# Study Hours by Gender
sns.boxplot(data=df, x='Gender', y='Study_Hours_Per_Week', ax=axes[1, 1], palette='Set2')
axes[1, 1].set_title('Study Hours Per Week by Gender')

plt.tight_layout()
plt.show()
```

```python
# Scatter Plots: Relationship with GPA
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Relationship with GPA', fontsize=16, fontweight='bold')

# Study Hours vs GPA
axes[0, 0].scatter(df['Study_Hours_Per_Week'], df['GPA'], alpha=0.6, color='steelblue', edgecolor='black')
axes[0, 0].set_xlabel('Study Hours Per Week')
axes[0, 0].set_ylabel('GPA')
axes[0, 0].set_title('Study Hours vs GPA')

# Attendance vs GPA
axes[0, 1].scatter(df['Attendance_Percentage'], df['GPA'], alpha=0.6, color='seagreen', edgecolor='black')
axes[0, 1].set_xlabel('Attendance %')
axes[0, 1].set_ylabel('GPA')
axes[0, 1].set_title('Attendance vs GPA')

# Average Score vs GPA
df['Avg_Score'] = (df['Math_Score'] + df['Science_Score'] + df['English_Score']) / 3
axes[1, 0].scatter(df['Avg_Score'], df['GPA'], alpha=0.6, color='coral', edgecolor='black')
axes[1, 0].set_xlabel('Average Subject Score')
axes[1, 0].set_ylabel('GPA')
axes[1, 0].set_title('Average Score vs GPA')

# Age vs GPA
axes[1, 1].scatter(df['Age'], df['GPA'], alpha=0.6, color='orchid', edgecolor='black')
axes[1, 1].set_xlabel('Age')
axes[1, 1].set_ylabel('GPA')
axes[1, 1].set_title('Age vs GPA')

plt.tight_layout()
plt.show()
```

```python
# Summary Statistics and Key Insights
print("\n" + "=" * 60)
print("KEY INSIGHTS FROM EDA")
print("=" * 60)
print(f"\n1. Total Students: {len(df)}")
print(f"2. Age Range: {df['Age'].min()} - {df['Age'].max()} years")
print(f"3. Average GPA: {df['GPA'].mean():.2f} (Std: {df['GPA'].std():.2f})")
print(f"4. Average Attendance: {df['Attendance_Percentage'].mean():.1f}%")
print(f"5. Average Study Hours/Week: {df['Study_Hours_Per_Week'].mean():.1f} hours")

print("\nScore Statistics:")
print(f"   - Math Score Average: {df['Math_Score'].mean():.1f}")
print(f"   - Science Score Average: {df['Science_Score'].mean():.1f}")
print(f"   - English Score Average: {df['English_Score'].mean():.1f}")

print("\nGender Distribution:")
for gender, count in df['Gender'].value_counts().items():
    percentage = (count / len(df)) * 100
    print(f"   - {gender}: {count} ({percentage:.1f}%)")

print("\nTop 5 Highest GPA Students:")
print(df.nlargest(5, 'GPA')[['Student_ID', 'Name', 'GPA', 'Attendance_Percentage', 'Study_Hours_Per_Week']])

print("\nBottom 5 Lowest GPA Students:")
print(df.nsmallest(5, 'GPA')[['Student_ID', 'Name', 'GPA', 'Attendance_Percentage', 'Study_Hours_Per_Week']])

print("\n" + "=" * 60)
```
